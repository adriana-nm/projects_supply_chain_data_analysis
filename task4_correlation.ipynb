```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Task 4: Correlation Analysis

```python
df = pd.read_csv('DataCoSupplyChainDataset.csv',encoding= 'unicode_escape')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

##### Create the correlation matrix & plot:

```python
np.round(df.corr(),2)
```

```python
plt.figure(figsize=(15,5)) 
sns.heatmap(np.round(df.corr(),2),annot=True)
```

##### Remove the NAN & duplicated columns and repeat the matrix & plot of correlation:

Remove the following NAN columns:  Product Description / Product Status

Remove the following duplicated columns:  Order Customer Id / Order Item Total / Benefit per Order / Order Item Cardprod Id / Product Category Id / Order Item Product Price

```python
df_corr = df[['Days for shipping (real)', 'Days for shipment (scheduled)', 'Sales per customer', 
       'Late_delivery_risk', 'Category Id', 'Customer Id', 'Customer Zipcode', 'Department Id', 
       'Latitude', 'Longitude','Order Id', 'Order Item Discount', 'Order Item Discount Rate', 
       'Order Item Id', 'Order Item Profit Ratio', 'Order Item Quantity', 'Sales',
       'Order Profit Per Order','Order Zipcode', 'Product Card Id', 'Product Price']]
np.round(df_corr.corr(),2)
```

```python
plt.figure(figsize=(15,10)) 
sns.heatmap(np.round(df_corr.corr(),2),annot=True)
```

##### Conclusion:

*There is a high positive correlation among the following variables:

    1 - Order Item Id & Order Id

    0.99 - Sales & Sales per Customer

    0.99 - Product Card Id & Category Id

    0.91 - Product Card Id & Department Id

    0.89 - Department ID & Category Id

    0.82 - Order Profit Per Order & Order Item Profit Ratio

    0.79 - Product Price & Sales

    0.78 - Product Price & Sales per customer

    0.66 - Order Discount Rate & Order Item Discount

*There is a high negative correlation among the following variables:

    -0.92 - Longitude & Customer Zip Code
