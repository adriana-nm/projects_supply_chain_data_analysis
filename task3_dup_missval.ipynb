```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Task 3: Dealing With Duplicate Rows and Missing Values

```python
df = pd.read_csv('DataCoSupplyChainDataset.csv',encoding= 'unicode_escape')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

##### Check if there are duplicates in the dataset:

```python
df.duplicated().sum()
```

##### Check how many values are missing in every column:

```python
df.isna().sum().sort_values(ascending=False)
```

##### Ratio (or %) of missing values from the total data:

```python
df.isna().sum()/len(df)
```

##### See the amount of nulls in a heatmap (in white):

```python
sns.heatmap(df.isnull(),cbar=False)
```

##### Conclusion:

* There aren't duplicate rows in the dataset


* The missing values are focused on two columns: Order Zip Code (86% of missing values) and Product Description (100% of missing values)


* Columns with missing values are not relevant in the analysis. These columns haven't been included in our previous analysis, so there's no need to recalculate our results.
